```python
import numpy as np
from PIL import Image
from wordcloud import WordCloud
```

```python
text=open(r'c.txt',encoding='utf-8').read()
```

```python
text
```

```python
img='images.jpg'
```

```python
mg=Image.open(img)
```

```python
mg
```

```python
Ip_mask=np.array(Image.open(img))
Ip_mask
```

```python
lp_wc=WordCloud(
    font_path='C:\Windows\Fonts\malgun.ttf',
    background_color='white',mask=Ip_mask,max_font_size=100,
                contour_width=5,contour_color='black').generate(text)
```

```python
lp_wc.to_image()
```
